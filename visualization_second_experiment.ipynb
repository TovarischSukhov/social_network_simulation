Frist, lets try run fairly small world - 100 workers and 10 employers

```python
exp = Experiment([0.5, 0.8, 0.74, 0.9, 0.95, 1.0],[0.5, 0.8, 0.74, 0.9, 0.95, 1.0])
exp.set_config({
        'no_workers': 100,
        'no_employers': 20,
        'no_vacancys_per_employer': 5,
        'no_connections': 5,
        'imposter_wage_coeff': 0.9,
    })
try:
    exp.run()
except KeyboardInterrupt:
    print(exp.current_world)
    exit(0)
```

```python
vlaues_to_plot = [(0.5, 0.5), (0.5, 0.8), (0.5, 0.95),
                  (0.8, 0.5), (0.8, 0.8), (0.8, 0.95),
                  (0.95, 0.5), (0.95, 0.8), (1.0, 1.0),
                 ]

def plot_result(exp, ax, alpha, betha):
    y = exp.results['al_' + str(alpha) + '_bet_' + str(betha)]
    x = np.arange(0, len(y))
    ax.plot(x, y, lw=1, color='blue')
```

```python
fig1, axs = plt.subplots(3, 3, figsize=(10, 8), constrained_layout=True)

axs = axs.flat
for ax, albt in zip(axs, vlaues_to_plot):
#     y = exp.results['al_' + str(alpha) + '_bet_' + str(betha)]
#     x = np.arange(0, len(y))
    ax.set_title(f'alpha={albt[0]}, betha={albt[1]}')
    plot_result(exp, ax, *albt)
```
